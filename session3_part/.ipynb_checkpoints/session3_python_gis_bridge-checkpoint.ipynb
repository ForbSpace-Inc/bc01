## Part 1: From Python to Spatial Data

### 1.1 WHAT IS SPATIAL DATA?
In Session #2, we worked with a table of airport data (names, elevations, traffic types). But we also had **coordinates** — latitude and longitude. Data that has a location is called **spatial data**.

There are two main types of spatial data:

| Type | Description | Example |
|---|---|---|
| **Vector Data** | Points, lines, and polygons | Airport locations, roads, country borders |
| **Raster Data** | A grid of pixels (like a photo) | Satellite images, elevation maps |

In this session, we'll work with both!

### 1.2 GEOJSON — A SPATIAL DATA FORMAT
In Session #2, we loaded an `aerodromes-ke.geojson` file. GeoJSON is a standard format for storing vector spatial data. Let's look at its structure.

````python
import json
import pandas as pd
import folium
import math
import numpy as np
import matplotlib.pyplot as plt

# Load the GeoJSON file
with open('aerodromes-ke.geojson', 'r', encoding='utf-8') as f:
    geojson_data = json.load(f)

# Inspect the structure of a single feature
first_feature = geojson_data['features'][0]
print("=== GeoJSON Feature Structure ===")
print(f"Type: {first_feature['type']}")
print(f"\nGeometry Type: {first_feature['geometry']['type']}")
print(f"Coordinates: {first_feature['geometry']['coordinates']}")
print(f"\nProperties (first 5):")
for key, value in list(first_feature['properties'].items())[:5]:
    print(f"  {key}: {value}")
````

````python
# Parse GeoJSON into a Pandas DataFrame (as we did in Session 2)
records = []
for feature in geojson_data['features']:
    props = feature['properties'].copy()
    coords = feature['geometry']['coordinates']
    props['longitude'] = coords[0]
    props['latitude']  = coords[1]
    records.append(props)

aerodromes = pd.DataFrame(records)
aerodromes.columns = [col.lower() for col in aerodromes.columns]
aerodromes = aerodromes.rename(columns={'icao_code': 'icao', 'traffic_permitted': 'traffic'})
aerodromes['elevation_ft'] = aerodromes['elevation_m_ft'].str.extract(r'([0-9.]+)\s*FT').astype(float)

print(f"✅ Loaded {len(aerodromes)} aerodromes with spatial coordinates")
aerodromes[['icao', 'traffic', 'elevation_ft', 'latitude', 'longitude']].head()
````

## Part 2: Interactive Maps with Folium

In Session #2, we created static charts with Matplotlib. Now let's create **interactive maps** using `folium` — a Python library that wraps the Leaflet.js mapping library.

Interactive maps allow users to zoom in, click on features, and explore the data dynamically.

### 2.1 CREATING A BASIC MAP

````python
# Create a base map centered on Kenya
# location=[latitude, longitude], zoom_start controls initial zoom level
m = folium.Map(
    location=[-1.286389, 36.817222],  # Nairobi coordinates
    zoom_start=6,
    tiles='OpenStreetMap'
)

# Display the map
print("✅ Base map created! (Scroll down to see it)")
m
````

### 2.2 ADDING AERODROME MARKERS
Let's add a marker for each aerodrome. Folium markers can have **popups** (click to see info) and **tooltips** (hover to see info).

````python
# Create a new map
m2 = folium.Map(location=[-1.286389, 36.817222], zoom_start=6)

# Add a marker for each aerodrome
for i, row in aerodromes.iterrows():
    if pd.notna(row['latitude']) and pd.notna(row['longitude']):
        # Create popup HTML content
        popup_html = f"""
        <div style="font-family: Arial; min-width: 150px;">
            <h4 style="margin: 0; color: #2c3e50;">{row['icao']}</h4>
            <hr style="margin: 5px 0;">
            <b>Elevation:</b> {row['elevation_ft']:.0f} ft<br>
            <b>Traffic:</b> {row['traffic']}<br>
            <b>Operator:</b> {row.get('admin_company', 'N/A')}<br>
            <b>Coordinates:</b> ({row['latitude']:.4f}, {row['longitude']:.4f})
        </div>
        """
        
        folium.Marker(
            location=[row['latitude'], row['longitude']],
            popup=folium.Popup(popup_html, max_width=250),
            tooltip=f"✈️ {row['icao']}",
            icon=folium.Icon(color='blue', icon='plane', prefix='fa')
        ).add_to(m2)

print("✅ Map with markers created! Click on a marker to see details.")
m2
````

### 2.3 COLOR-CODING BY TRAFFIC TYPE
Let's make our map more informative by using different colors for IFR/VFR and VFR-only aerodromes.

````python
# Create a color-coded map
m3 = folium.Map(location=[-1.286389, 36.817222], zoom_start=6)

# Define colors for each traffic type
def get_marker_color(traffic_type):
    if 'IFR' in str(traffic_type):
        return 'red'    # IFR/VFR airports in red
    else:
        return 'green'  # VFR-only airports in green

# Add markers with appropriate colors
for i, row in aerodromes.iterrows():
    if pd.notna(row['latitude']) and pd.notna(row['longitude']):
        color = get_marker_color(row['traffic'])
        
        popup_html = f"""
        <div style="font-family: Arial; min-width: 150px;">
            <h4 style="margin: 0; color: #2c3e50;">{row['icao']}</h4>
            <hr style="margin: 5px 0;">
            <b>Traffic:</b> {row['traffic']}<br>
            <b>Elevation:</b> {row['elevation_ft']:.0f} ft
        </div>
        """
        
        folium.CircleMarker(
            location=[row['latitude'], row['longitude']],
            radius=8,
            popup=folium.Popup(popup_html, max_width=200),
            tooltip=f"{row['icao']} — {row['traffic']}",
            color='black',
            weight=1,
            fill=True,
            fill_color=color,
            fill_opacity=0.8
        ).add_to(m3)

# Add a legend
legend_html = """
<div style="position: fixed; bottom: 30px; left: 30px; z-index: 1000; 
            background-color: white; padding: 10px; border-radius: 5px;
            border: 2px solid grey; font-family: Arial;">
    <b>Traffic Type</b><br>
    <span style="color: red;">●</span> IFR/VFR<br>
    <span style="color: green;">●</span> VFR only
</div>
"""
m3.get_root().html.add_child(folium.Element(legend_html))

print("✅ Color-coded map created!")
m3
````

## Part 3: Spatial Analysis with Python

GIS is not just about making maps — it's about **analyzing spatial relationships**. Let's use Python to perform spatial analysis on our aerodrome data.

### 3.1 NEAREST NEIGHBOR ANALYSIS
Which aerodrome is most isolated? Which ones are clustered together?

We'll use a data structure called a **KD-Tree** (K-Dimensional Tree) for efficient nearest-neighbor searches.

````python
from scipy.spatial import KDTree

# Extract coordinates (dropping any rows with missing coordinates)
valid_aerodromes = aerodromes[['icao', 'latitude', 'longitude']].dropna()
coords = valid_aerodromes[['latitude', 'longitude']].values
icaos = valid_aerodromes['icao'].values

# Build a KD-Tree for fast nearest-neighbor searches
tree = KDTree(coords)

# For each aerodrome, find its nearest neighbor
# k=2 because the nearest neighbor to itself is itself (distance 0)
distances, indices = tree.query(coords, k=2)

# The nearest neighbor distance is the second result (index 1)
nearest_distances_deg = distances[:, 1]
nearest_distances_km = nearest_distances_deg * 111  # Approximate: 1 degree ≈ 111 km

isolation_df = pd.DataFrame({
    'icao': icaos,
    'nearest_neighbor_km': nearest_distances_km.round(1)
}).sort_values('nearest_neighbor_km', ascending=False)

print("=== TOP 10 MOST ISOLATED AERODROMES ===")
print("(Furthest from their nearest neighbor)")
print(isolation_df.head(10).to_string(index=False))
````

### 3.2 DISTANCE ANALYSIS FROM JKIA
Let's calculate the distance from Jomo Kenyatta International Airport (HKJK) to all other aerodromes and visualize it on a map.

````python
def haversine_distance(lat1, lon1, lat2, lon2):
    """Great-circle distance in km between two lat/lon points."""
    R = 6371
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    dlat, dlon = lat2 - lat1, lon2 - lon1
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    return R * 2 * math.asin(math.sqrt(a))

# Get JKIA coordinates
jkia = aerodromes[aerodromes['icao'] == 'HKJK'].iloc[0]
jkia_lat, jkia_lon = jkia['latitude'], jkia['longitude']

# Calculate distances
distances = []
for _, row in aerodromes.iterrows():
    if row['icao'] != 'HKJK' and pd.notna(row['latitude']):
        dist = haversine_distance(jkia_lat, jkia_lon, row['latitude'], row['longitude'])
        distances.append({'icao': row['icao'], 'distance_km': round(dist, 1), 'traffic': row['traffic']})

distance_df = pd.DataFrame(distances).sort_values('distance_km')
print("Aerodromes sorted by distance from JKIA:")
print(distance_df.to_string(index=False))
````

````python
# Visualize distances on a map with lines from JKIA
m_dist = folium.Map(location=[-1.286389, 36.817222], zoom_start=6)

# Add JKIA marker (special)
folium.Marker(
    location=[jkia_lat, jkia_lon],
    popup="JKIA — Reference Point",
    tooltip="HKJK (JKIA)",
    icon=folium.Icon(color='red', icon='star', prefix='fa')
).add_to(m_dist)

# Add other aerodromes with lines to JKIA
for _, row in aerodromes.iterrows():
    if row['icao'] != 'HKJK' and pd.notna(row['latitude']):
        dist = haversine_distance(jkia_lat, jkia_lon, row['latitude'], row['longitude'])
        
        # Color based on distance
        if dist < 100:
            color = '#2ecc71'  # Green = close
        elif dist < 300:
            color = '#f39c12'  # Orange = medium
        else:
            color = '#e74c3c'  # Red = far
        
        # Draw line from JKIA to this aerodrome
        folium.PolyLine(
            locations=[[jkia_lat, jkia_lon], [row['latitude'], row['longitude']]],
            color=color, weight=1.5, opacity=0.6
        ).add_to(m_dist)
        
        # Add marker
        folium.CircleMarker(
            location=[row['latitude'], row['longitude']],
            radius=6,
            popup=f"{row['icao']}: {dist:.0f} km from JKIA",
            tooltip=f"{row['icao']} — {dist:.0f} km",
            color='black', weight=1, fill=True, fill_color=color, fill_opacity=0.8
        ).add_to(m_dist)

print("✅ Distance map created!")
m_dist
````

## Part 4: Spatial Clustering with Machine Learning

Machine learning can help us find patterns in spatial data. **DBSCAN** (Density-Based Spatial Clustering of Applications with Noise) is an algorithm that groups points that are close together into clusters.

This is used in GIS for tasks like:
- Finding urban clusters
- Identifying hotspots of activity
- Grouping nearby infrastructure

### 4.1 RUNNING DBSCAN CLUSTERING

````python
from sklearn.cluster import DBSCAN

# Extract coordinates for clustering
coords_cluster = aerodromes[['latitude', 'longitude']].dropna().values

# Run DBSCAN
# eps: maximum distance between two points to be in the same cluster (in degrees, ~0.5° ≈ 55 km)
# min_samples: minimum number of points to form a cluster
clustering = DBSCAN(eps=0.5, min_samples=2).fit(coords_cluster)

# Add cluster labels to our DataFrame
aerodromes_clustered = aerodromes.loc[aerodromes[['latitude', 'longitude']].dropna().index].copy()
aerodromes_clustered['cluster'] = clustering.labels_

# Summary
n_clusters = len(set(clustering.labels_)) - (1 if -1 in clustering.labels_ else 0)
n_isolated = list(clustering.labels_).count(-1)

print(f"✅ DBSCAN Analysis Complete!")
print(f"   Found {n_clusters} cluster(s) of aerodromes")
print(f"   {n_isolated} aerodrome(s) are isolated (not in any cluster)")

# Show which aerodromes are in each cluster
print("\n=== Aerodromes by Cluster ===")
for cluster_id in sorted(set(clustering.labels_)):
    label = 'Isolated' if cluster_id == -1 else f'Cluster {cluster_id}'
    members = aerodromes_clustered[aerodromes_clustered['cluster'] == cluster_id]
    print(f"\n{label} ({len(members)} aerodromes):")
    for _, row in members.iterrows():
        print(f"  - {row['icao']}")
````

### 4.2 VISUALIZING CLUSTERS ON A MAP

````python
# Create a cluster visualization map
m_cluster = folium.Map(location=[-1.286389, 36.817222], zoom_start=6)

# Define colors for each cluster
cluster_colors = {
    -1: 'gray',   # Isolated
    0: 'red',
    1: 'blue',
    2: 'purple',
    3: 'orange',
    4: 'darkgreen'
}

for _, row in aerodromes_clustered.iterrows():
    if pd.notna(row['latitude']) and pd.notna(row['longitude']):
        cluster_id = row['cluster']
        color = cluster_colors.get(cluster_id, 'black')
        
        cluster_label = 'Isolated' if cluster_id == -1 else f'Cluster {cluster_id}'
        
        folium.CircleMarker(
            location=[row['latitude'], row['longitude']],
            radius=10,
            popup=f"{row['icao']} — {cluster_label}",
            tooltip=f"{row['icao']} ({cluster_label})",
            color='black', weight=1, fill=True, fill_color=color, fill_opacity=0.8
        ).add_to(m_cluster)

# Add legend
legend_html = """
<div style="position: fixed; bottom: 30px; left: 30px; z-index: 1000; 
            background-color: white; padding: 10px; border-radius: 5px;
            border: 2px solid grey; font-family: Arial;">
    <b>DBSCAN Clusters</b><br>
    <span style="color: gray;">●</span> Isolated<br>
    <span style="color: red;">●</span> Cluster 0<br>
    <span style="color: blue;">●</span> Cluster 1<br>
    <span style="color: purple;">●</span> Cluster 2
</div>
"""
m_cluster.get_root().html.add_child(folium.Element(legend_html))

print("✅ Cluster map created!")
m_cluster
````

## Part 5: Introduction to Remote Sensing with Python

Your colleague will cover the theoretical foundations of Remote Sensing. This section is your **Python bridge** to those concepts — showing you how Python handles satellite imagery.

### 5.1 WHAT IS RASTER DATA?
Satellite images are stored as **raster data** — a grid of pixels, where each pixel has a value. Unlike our aerodrome points (vector data), raster data covers an entire area.

```
Vector Data (Points):     Raster Data (Grid):
    •  •                  [120][130][140][150]
  •      •                [110][125][145][160]
    •  •                  [100][115][135][155]
                          [90 ][105][125][145]
```

### 5.2 SATELLITE BANDS
Satellites capture images in multiple **bands** (wavelengths of light). Different bands reveal different information:

| Band | Wavelength | What It Shows |
|---|---|---|
| Blue | 450-520 nm | Water bodies, atmosphere |
| Green | 520-600 nm | Vegetation, water |
| Red | 630-690 nm | Vegetation, soil |
| Near-Infrared (NIR) | 760-900 nm | Vegetation health |
| Short-Wave Infrared | 1550-1750 nm | Soil moisture, geology |

### 5.4 NDVI — MEASURING VEGETATION HEALTH
**NDVI** (Normalized Difference Vegetation Index) is one of the most widely used remote sensing indices. It measures how healthy and dense vegetation is.

**Formula:** `NDVI = (NIR - Red) / (NIR + Red)`

- **NDVI close to +1:** Dense, healthy vegetation (forests)
- **NDVI close to 0:** Sparse vegetation or bare soil
- **NDVI close to -1:** Water bodies

This is exactly what we'll use to analyze Mt. Elgon!

## Part 6: Coordinate Reference Systems (CRS)

One of the most important concepts in GIS is the **Coordinate Reference System (CRS)**. It defines how coordinates relate to actual locations on Earth.

### 6.1 WHY CRS MATTERS
The Earth is a 3D sphere, but maps are 2D. When we "flatten" the Earth onto a map, we need to choose a **projection**. Different projections distort different things (shape, area, distance, direction).

| CRS | EPSG Code | Description |
|---|---|---|
| **WGS84** | EPSG:4326 | Standard GPS coordinates (lat/lon in degrees) |
| **Web Mercator** | EPSG:3857 | Used by Google Maps, OpenStreetMap |
| **UTM Zone 37N** | EPSG:32637 | Good for Kenya (minimizes distortion) |

### 6.2 WORKING WITH CRS IN PYTHON

````python
# Demonstrate coordinate transformation concept
# (In practice, you'd use pyproj or geopandas for this)

# WGS84 coordinates of JKIA
jkia_lat_wgs84 = -1.3192
jkia_lon_wgs84 = 36.9278

print("=== Coordinate Reference Systems ===")
print(f"\nJKIA in WGS84 (EPSG:4326):")
print(f"  Latitude:  {jkia_lat_wgs84}°")
print(f"  Longitude: {jkia_lon_wgs84}°")

# Approximate conversion to Web Mercator (EPSG:3857)
# Formula: x = lon * 20037508.34 / 180
#          y = log(tan((90 + lat) * pi / 360)) * 20037508.34 / pi
import math
x_mercator = jkia_lon_wgs84 * 20037508.34 / 180
y_mercator = math.log(math.tan((90 + jkia_lat_wgs84) * math.pi / 360)) * 20037508.34 / math.pi

print(f"\nJKIA in Web Mercator (EPSG:3857) — approximate:")
print(f"  X: {x_mercator:.2f} meters")
print(f"  Y: {y_mercator:.2f} meters")

print("\n📌 In the Case Study, GeoPandas handles CRS transformations automatically!")
````
