## Part 1: Python Basics — The Building Blocks

Before we analyze real airport data, let's learn the fundamental building blocks of Python. We'll use aviation examples to make everything relatable!

### 1.1 VARIABLES — Storing Information
A **variable** is like a labeled box where you store information. You give the box a name, and you can put any kind of data inside it.

**Syntax:** `variable_name = value`

````python
# Think of variables as labeled boxes for storing data
airport_name = "Jomo Kenyatta International"  # A text (string)
icao_code = "HKJK"                            # Another text
elevation_feet = 5330                          # A whole number (integer)
has_control_tower = True                       # True or False (boolean)
runway_length_km = 4.1                         # A decimal number (float)

# The print() function displays output on screen
print("Airport:", airport_name)
print("ICAO Code:", icao_code)
print("Elevation:", elevation_feet, "feet above sea level")
print("Has Control Tower:", has_control_tower)
print("Runway Length:", runway_length_km, "km")
````

### 1.2 DATA TYPES — Different Kinds of Information
Python has different types of data. The main ones are:

| Type | Description | Example |
|---|---|---|
| `str` (String) | Text | `"Nairobi"`, `"HKJK"` |
| `int` (Integer) | Whole numbers | `5330`, `2`, `100` |
| `float` (Float) | Decimal numbers | `3.14`, `-1.3192` |
| `bool` (Boolean) | True or False | `True`, `False` |

````python
# Strings (text) - Airport names, codes
airport_code = "HKNW"
print(f"airport_code is a: {type(airport_code)}")

# Integers (whole numbers) - Elevation, runway counts
runway_count = 2
print(f"runway_count is a: {type(runway_count)}")

# Floats (decimal numbers) - Coordinates, distances
latitude = -1.3217
longitude = 36.8148
print(f"latitude is a: {type(latitude)}")

# Booleans (True/False) - Yes/No questions
has_ils = True  # ILS = Instrument Landing System
print(f"has_ils is a: {type(has_ils)}")

# f-strings: A convenient way to include variables in text
print(f"\nWilson Airport is at coordinates: ({latitude}, {longitude})")
````

### 1.3 ARITHMETIC OPERATIONS — Python as a Calculator
Python can do all kinds of math. This is very useful in aeronautics!

````python
# Basic arithmetic
print("Basic Arithmetic:")
print(f"  10 + 5 = {10 + 5}")
print(f"  10 - 5 = {10 - 5}")
print(f"  10 * 5 = {10 * 5}")
print(f"  10 / 3 = {10 / 3:.2f}")  # .2f means 2 decimal places
print(f"  10 ** 2 = {10 ** 2}")    # ** means "to the power of"

# Aviation example: Converting units
print("\n--- Aviation Unit Conversions ---")
speed_knots = 450
speed_kmh = speed_knots * 1.852
print(f"  {speed_knots} knots = {speed_kmh:.1f} km/h")

fuel_liters = 50000
fuel_kg = fuel_liters * 0.8  # Jet fuel density ~0.8 kg/L
print(f"  {fuel_liters} liters of jet fuel = {fuel_kg:.0f} kg")
````

### 1.4 LISTS — Collections of Items

### 1.5 DICTIONARIES — Storing Related Information Together

### 1.6 IF STATEMENTS — Making Decisions
An **if statement** lets your code make decisions based on conditions.

**Syntax:**
```python
if condition:
    # do this if condition is True
else:
    # do this if condition is False
```

````python
# Check if an airport is at high altitude
elevation = 5330  # JKIA elevation in feet

if elevation > 5000:
    print(f"JKIA is a HIGH ALTITUDE airport at {elevation} ft.")
    print("Pilots need to adjust their calculations for thinner air!")
else:
    print(f"JKIA is at a normal altitude of {elevation} ft.")

# Multiple conditions with elif
print("\n--- Traffic Type Classification ---")
traffic = "IFR/VFR"

if traffic == "IFR/VFR":
    print("This airport handles both Instrument and Visual Flight Rules traffic.")
elif traffic == "VFR":
    print("This airport only handles Visual Flight Rules traffic.")
else:
    print("Unknown traffic type.")
````

### 1.7 FOR LOOPS — Repeating Actions
A **for loop** lets you repeat the same action for every item in a list. Imagine checking the status of all 19 Kenyan airports — a loop does this automatically!

````python
airports_list = [
    {"name": "Jomo Kenyatta Intl",  "icao": "HKJK", "elevation": 5330},
    {"name": "Wilson Airport",       "icao": "HKNW", "elevation": 5546},
    {"name": "Kisumu Intl",          "icao": "HKKI", "elevation": 3795},
    {"name": "Moi Intl (Mombasa)",   "icao": "HKMO", "elevation": 200},
    {"name": "Eldoret Intl",         "icao": "HKEL", "elevation": 6945}
]

print("=== All Airports ===")
for airport in airports_list:
    print(f"  {airport['icao']}: {airport['name']} — {airport['elevation']} ft")

print("\n=== High Altitude Airports (>5000 ft) ===")
for airport in airports_list:
    if airport['elevation'] > 5000:
        print(f"  ✈️  {airport['name']} at {airport['elevation']} ft")
````

### 1.8 WHILE LOOPS — Repeating Until a Condition Changes
A **while loop** keeps running as long as a condition is True. This is useful for simulations, like monitoring an aircraft's descent.

````python
# Simulating an aircraft descent
altitude = 10000  # feet
target_altitude = 0
descent_rate = 1000  # feet per step

print("=== Aircraft Descent Simulation ===")
print(f"Starting at {altitude} ft. Descending...")

while altitude > target_altitude:
    altitude -= descent_rate
    
    if altitude > 3000:
        status = "En route"
    elif altitude > 1000:
        status = "⚠️  Entering traffic pattern"
    else:
        status = "🛬 On final approach"
    
    print(f"  Altitude: {altitude:,} ft — {status}")

print("\n✅ Landed safely!")
````

### 1.9 FUNCTIONS — Reusable Code Blocks
A **function** is a named block of code that you can call (use) multiple times. This saves you from writing the same code over and over.

**Syntax:**
```python
def function_name(input_parameter):
    # code to run
    return result
```

````python
# Define reusable unit conversion functions
def meters_to_feet(meters):
    """Convert meters to feet. Aviation uses feet for altitude!"""
    return meters * 3.28084

def feet_to_meters(feet):
    """Convert feet to meters."""
    return feet / 3.28084

def knots_to_kmh(knots):
    """Convert speed from knots to km/h."""
    return knots * 1.852

def celsius_to_fahrenheit(celsius):
    """Convert temperature from Celsius to Fahrenheit."""
    return (celsius * 9/5) + 32

# Use the functions
runway_length_m = 4100  # JKIA main runway in meters
print(f"JKIA runway: {runway_length_m}m = {meters_to_feet(runway_length_m):.0f} ft")

cruise_speed_knots = 450
print(f"Cruise speed: {cruise_speed_knots} knots = {knots_to_kmh(cruise_speed_knots):.1f} km/h")

temp_celsius = 25
print(f"Temperature: {temp_celsius}°C = {celsius_to_fahrenheit(temp_celsius):.1f}°F")
````

## Part 2: Python in Aeronautics — Working with Real Data

Now that we know the basics, let's apply them to **real aerodrome data** from the Kenya Aeronautical Information Publication (AIP). This is the official document that contains information about all airports in Kenya!

### 2.1 IMPORTING PYTHON LIBRARIES
Python has thousands of pre-built libraries (collections of code) that add extra capabilities. We need to import the ones we'll use.

````python
# Import the libraries we need
import json           # For reading JSON files
import pandas as pd   # For data analysis (like Excel but in Python)
import matplotlib.pyplot as plt  # For creating charts
import numpy as np    # For mathematical operations
import math           # For trigonometry (needed for distance calculations)

# Set a nice style for our charts
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline

print("✅ All libraries imported successfully!")
print(f"   Pandas version: {pd.__version__}")
print(f"   NumPy version: {np.__version__}")
````

### 2.2 LOADING THE AERODROMES DATA
Our data is stored in a **GeoJSON** file. GeoJSON is a standard format for storing geographic data. Each aerodrome is a "feature" with properties (like name, elevation) and geometry (coordinates).

````python
# Load the GeoJSON file
with open('../aerodromes-ke.geojson', 'r', encoding='utf-8') as f:
    geojson_data = json.load(f)

# Parse each feature into a flat record
records = []
for feature in geojson_data['features']:
    props = feature['properties'].copy()
    coords = feature['geometry']['coordinates']  # [longitude, latitude]
    props['longitude'] = coords[0]
    props['latitude']  = coords[1]
    records.append(props)

# Create a Pandas DataFrame (like a spreadsheet in Python)
aerodromes_raw = pd.DataFrame(records)

print(f"✅ Loaded {len(aerodromes_raw)} aerodromes from the Kenya AIP")
print(f"   Columns: {list(aerodromes_raw.columns)}")
aerodromes_raw.head()
````

### 2.3 CLEANING & PREPARING THE DATA
Real-world data is often messy. We need to select the columns we care about, rename them, and extract the numeric elevation value.

````python
# Select and rename the columns we need
aerodromes = aerodromes_raw[[
    'ICAO_Code', 'Type', 'Traffic_Permitted', 'Admin_Company',
    'Elevation_m_ft', 'Remarks_Spatial', 'longitude', 'latitude'
]].copy()

aerodromes.columns = [
    'icao', 'type', 'traffic', 'operator',
    'elevation_raw', 'remarks', 'longitude', 'latitude'
]

# Extract the numeric elevation value from the text (e.g., "5330 FT (1625 M)" → 5330)
aerodromes['elevation_ft'] = (
    aerodromes['elevation_raw']
    .str.extract(r'([0-9.]+)\s*FT')
    .astype(float)
)

print("✅ Data cleaned and ready!")
print(f"   Shape: {aerodromes.shape[0]} rows × {aerodromes.shape[1]} columns")
aerodromes[['icao', 'traffic', 'elevation_ft', 'latitude', 'longitude']].head(10)
````

### 2.4 EXPLORING THE DATASET
Let's get to know our data better with some basic statistics.

````python
# Basic statistics about our dataset
print("=== Dataset Overview ===")
print(f"Total aerodromes: {len(aerodromes)}")
print(f"\nElevation Statistics (feet):")
print(f"  Highest: {aerodromes['elevation_ft'].max():.0f} ft — {aerodromes.loc[aerodromes['elevation_ft'].idxmax(), 'icao']}")
print(f"  Lowest:  {aerodromes['elevation_ft'].min():.0f} ft — {aerodromes.loc[aerodromes['elevation_ft'].idxmin(), 'icao']}")
print(f"  Average: {aerodromes['elevation_ft'].mean():.0f} ft")

print(f"\nTraffic Types:")
print(aerodromes['traffic'].value_counts().to_string())

print(f"\nAerodrome Operators:")
print(aerodromes['operator'].value_counts().to_string())
````

### 2.5 FILTERING & QUERYING THE DATA
We can filter our DataFrame to find specific aerodromes, just like using a search filter.

````python
# Filter: Aerodromes that support IFR (Instrument Flight Rules)
# IFR means pilots can fly in clouds and low visibility using instruments
ifr_airports = aerodromes[aerodromes['traffic'].str.contains('IFR', na=False)]
print(f"IFR-capable aerodromes: {len(ifr_airports)}")
print(ifr_airports[['icao', 'traffic', 'elevation_ft']].to_string())

print("\n" + "="*50)

# Filter: High elevation aerodromes (> 5000 ft)
# These require special pilot training and aircraft performance calculations
high_altitude = aerodromes[aerodromes['elevation_ft'] > 5000].sort_values('elevation_ft', ascending=False)
print(f"\nHigh altitude aerodromes (>5000 ft): {len(high_altitude)}")
print(high_altitude[['icao', 'elevation_ft']].to_string())
````

## Part 3: Aeronautics Calculations with Python

Let's use Python to perform real calculations that pilots and flight planners use every day!

### 3.1 HAVERSINE DISTANCE FORMULA
The Earth is a sphere, so we can't use a simple straight-line formula to calculate the distance between two airports. We use the **Haversine formula**, which accounts for the Earth's curvature.

This is the same formula used in aviation navigation systems!

````python
def haversine_distance(lat1, lon1, lat2, lon2):
    """
    Calculate the great-circle distance between two points on Earth.
    
    Parameters:
        lat1, lon1: Coordinates of the first point (in decimal degrees)
        lat2, lon2: Coordinates of the second point (in decimal degrees)
    
    Returns:
        Distance in kilometers
    """
    R = 6371  # Earth's radius in kilometers
    
    # Convert degrees to radians (math functions need radians)
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    
    return R * c

# Test: Distance from JKIA (Nairobi) to Moi Intl (Mombasa)
jkia_lat, jkia_lon = -1.3192, 36.9278  # Nairobi
moi_lat, moi_lon   = -4.0348, 39.5942  # Mombasa

dist_km = haversine_distance(jkia_lat, jkia_lon, moi_lat, moi_lon)
dist_nm = dist_km / 1.852  # Convert to nautical miles (used in aviation)

print(f"Distance from JKIA (Nairobi) to Moi Intl (Mombasa):")
print(f"  {dist_km:.1f} km")
print(f"  {dist_nm:.1f} nautical miles")
````

### 3.2 BEARING (DIRECTION) CALCULATION
A **bearing** tells a pilot which direction to fly. 0° is North, 90° is East, 180° is South, 270° is West.

````python
def calculate_bearing(lat1, lon1, lat2, lon2):
    """
    Calculate the initial bearing (direction) from point 1 to point 2.
    
    Returns:
        Bearing in degrees (0-360), where 0 = North, 90 = East, etc.
    """
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    
    x = math.sin(dlon) * math.cos(lat2)
    y = math.cos(lat1) * math.sin(lat2) - math.sin(lat1) * math.cos(lat2) * math.cos(dlon)
    
    bearing = (math.degrees(math.atan2(x, y)) + 360) % 360
    return bearing

# Bearing from JKIA to Moi Intl
bearing = calculate_bearing(jkia_lat, jkia_lon, moi_lat, moi_lon)
print(f"Initial bearing from JKIA to Moi Intl: {bearing:.1f}°")
print("(Roughly South-East, which makes sense geographically!)")

# Bearing from JKIA to Wilson Airport (just across Nairobi)
wilson_lat, wilson_lon = -1.3213, 36.8150
bearing_wilson = calculate_bearing(jkia_lat, jkia_lon, wilson_lat, wilson_lon)
print(f"\nBearing from JKIA to Wilson Airport: {bearing_wilson:.1f}°")
````

### 3.3 DISTANCES FROM JKIA TO ALL AERODROMES
Let's calculate the distance from JKIA to every other aerodrome in Kenya.

````python
# Get JKIA's coordinates from our dataset
jkia_row = aerodromes[aerodromes['icao'] == 'HKJK'].iloc[0]
jkia_lat = jkia_row['latitude']
jkia_lon = jkia_row['longitude']

# Calculate distance from JKIA to every other aerodrome
distances = []
for _, row in aerodromes.iterrows():
    if row['icao'] != 'HKJK' and pd.notna(row['latitude']):
        dist = haversine_distance(jkia_lat, jkia_lon, row['latitude'], row['longitude'])
        distances.append({
            'icao': row['icao'],
            'distance_km': round(dist, 1),
            'distance_nm': round(dist / 1.852, 1),
            'traffic': row['traffic']
        })

distance_df = pd.DataFrame(distances).sort_values('distance_km')
print("Aerodromes sorted by distance from JKIA (Nairobi):")
print(distance_df.to_string(index=False))
````

## Part 4: Visualizing Aerodrome Data with Charts

Data is much easier to understand when we visualize it. Let's create some charts using `matplotlib`.

### 4.1 BAR CHART — Elevation Comparison

````python
# Sort aerodromes by elevation (highest first)
sorted_aerodromes = aerodromes.sort_values('elevation_ft', ascending=False)

fig, ax = plt.subplots(figsize=(12, 6))

# Create bar chart with color-coding by elevation
colors = ['#e74c3c' if e > 5000 else '#3498db' for e in sorted_aerodromes['elevation_ft']]
bars = ax.bar(sorted_aerodromes['icao'], sorted_aerodromes['elevation_ft'], 
              color=colors, edgecolor='black', linewidth=0.5)

# Add a horizontal line at 5000 ft
ax.axhline(y=5000, color='red', linestyle='--', linewidth=1.5, label='5,000 ft threshold')

ax.set_title('Elevation of Kenyan Aerodromes', fontsize=15, fontweight='bold', pad=15)
ax.set_xlabel('ICAO Code', fontsize=12)
ax.set_ylabel('Elevation (Feet)', fontsize=12)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height:.0f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3), textcoords="offset points",
                ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
````

### 4.2 SCATTER PLOT — Geographic Distribution
Let's plot the coordinates of our aerodromes to see how they are spread across Kenya.

````python
fig, ax = plt.subplots(figsize=(10, 9))

# Color-code by traffic type
traffic_colors = {'IFR/VFR': '#e74c3c', 'VFR': '#2ecc71'}
for traffic_type, color in traffic_colors.items():
    subset = aerodromes[aerodromes['traffic'] == traffic_type]
    ax.scatter(subset['longitude'], subset['latitude'],
               label=traffic_type, c=color, s=100, alpha=0.8,
               edgecolors='black', linewidth=0.7, zorder=5)

# Add ICAO code labels
for _, row in aerodromes.iterrows():
    if pd.notna(row['longitude']) and pd.notna(row['latitude']):
        ax.annotate(row['icao'], (row['longitude'], row['latitude']),
                    xytext=(5, 5), textcoords='offset points', fontsize=8, zorder=6)

ax.set_title('Geographic Distribution of Kenyan Aerodromes', fontsize=14, fontweight='bold')
ax.set_xlabel('Longitude (°E)', fontsize=12)
ax.set_ylabel('Latitude (°N/S)', fontsize=12)
ax.legend(title='Traffic Type', bbox_to_anchor=(1.02, 1), loc='upper left')
ax.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
````
