```python
aircraft_speeds = [850, 920, 780, 1020, 890]

print("✈️ Aircraft Speed Analysis")

for speed in aircraft_speeds:
    if speed > 900:
        print(speed, "km/h → High-speed aircraft")
    else:
        print(speed, "km/h → Normal speed")
```

```python
speeds = [850, 920, 780, 1020, 890]

average_speed = sum(speeds) / len(speeds)

print("Average aircraft speed:", average_speed, "km/h")
```

### Rocket Launch Countdown with System Checks

```python
import time

systems = ["Fuel System", "Navigation", "Communication", "Engine"]

print("🚀 Pre-launch System Check")

for system in systems:
    print("Checking", system, "...")
    time.sleep(1)
    print(system, "OK")

print("\nStarting Launch Countdown")

for i in range(10, 0, -1):
    print(i)
    time.sleep(1)

print("🚀 Launch Successful!")
```

### Satellite Orbit Calculator 🛰️

```python
altitude = int(input("Enter satellite altitude (km): "))

if altitude < 200:
    print("Too low: satellite will burn in atmosphere")
elif altitude < 2000:
    print("Low Earth Orbit (LEO)")
elif altitude < 35786:
    print("Medium Earth Orbit (MEO)")
else:
    print("Geostationary Orbit")
```

### Mini Project: Aircraft Flight Tracker ✈️

```python
aircraft = []

print("✈️ Aircraft Tracker")

for i in range(3):
    name = input("Enter aircraft name: ")
    altitude = input("Enter altitude (m): ")
    aircraft.append((name, altitude))

print("\nActive Aircraft:")

for plane in aircraft:
    print("Aircraft:", plane[0], "| Altitude:", plane[1], "m")
```
