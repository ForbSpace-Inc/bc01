## 2. Conditionals: Making Ship Decisions

Should we launch? Should we abort? Space flight requires constant checks. We use `if` statements to handle these decisions.

```python
fuel = 85
weather_clear = True

print("Checking pre-launch status...")

if fuel >= 80 and weather_clear:
    print("🚀 Status: Ready for Launch!")
elif fuel < 80:
    print("⚠️ Status: Warning - Low Fuel!")
else:
    print("❌ Status: Abort - Bad Weather.")
```

## 3. Loops: Repetitive Tasks (Countdown & Scanning)

Computers are great at doing things over and over again without getting bored. Like counting down to zero or scanning a list of nearby planets.

```python
import time

print("Starting Countdown...")

for i in range(10, 0, -1):
    print(f"{i}...")
    time.sleep(0.5) # Wait half a second

print("🚀 LIFT OFF!")
```

### Scanning the Star Map
We can use a `list` to store many items, and a `for` loop to check each one.

```python
planets = ["Mercury", "Venus", "Earth", "Mars", "Jupiter"]

print("Scanning list of nearby celestial bodies:")
for planet in planets:
    if planet == "Mars":
        print(f"  -> {planet} FOUND! (Target Destination)")
    else:
        print(f"  -> Scanning {planet}...")
```

---
## 4. Functions: Reusable Tools

In space travel, we often need to calculate things like distance. Instead of writing the math every time, we build a **function**.

```python
def calculate_distance(speed, time_hours):
    """Calculates distance traveled in space"""
    return speed * time_hours

# A rocket traveling at 28,000 km/h for 24 hours
total_dist = calculate_distance(28000, 24)
print(f"Total distance traveled in one day: {total_dist:,} km")
```
