### 1.8 A GIS-SPECIFIC FUNCTION - Bearing between two aerodromes


```python
import math
import folium

def calculate_bearing(lat1, lon1, lat2, lon2):
    """Calculate bearing (direction) between two points."""
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    x = math.sin(dlon) * math.cos(lat2)
    y = math.cos(lat1) * math.sin(lat2) - math.sin(lat1) * math.cos(lat2) * math.cos(dlon)
    bearing = (math.degrees(math.atan2(x, y)) + 360) % 360
    return bearing

# Bearing from JKIA to Wilson Airport
bearing = calculate_bearing(-1.3192, 36.9278, -1.3213, 36.8150)
print(f"Bearing from JKIA to Wilson: {bearing:.1f}°")
```

## Part 2: Loading Real Aerodromes Data

Now we're ready to work with actual aerodromes data from the **Kenya AIP GeoJSON** file. We'll parse it into a Pandas DataFrame and explore its structure.

### 2.1 IMPORTING OUR TOOLS


```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline

print("✅ Libraries imported successfully!")
```

### 2.2 LOADING THE AERODROMES GEOJSON


```python
# The GeoJSON file contains aerodrome points from the Kenya AIP.
# Each feature has 'properties' (attributes) and 'geometry' (coordinates).

with open('aerodromes-ke.geojson', 'r', encoding='utf-8') as f:
    geojson_data = json.load(f)

# Parse features into a flat list of records
records = []
for feature in geojson_data['features']:
    props = feature['properties'].copy()
    coords = feature['geometry']['coordinates']  # [longitude, latitude]
    props['longitude'] = coords[0]
    props['latitude']  = coords[1]
    records.append(props)

aerodromes_raw = pd.DataFrame(records)

print(f"✅ Loaded {len(aerodromes_raw)} aerodromes")
aerodromes_raw.head()
```

### 2.3 CLEANING & PREPARING THE DATA


```python
# The Elevation_m_ft column stores values like "5330 FT (1625 M)" or "1690 / 5546".
# We'll extract just the feet value.

import re

def parse_elevation_ft(raw):
    """Extract the first numeric value from an elevation string (feet)."""
    if pd.isna(raw):
        return np.nan
    # Match the first integer or decimal number in the string
    match = re.match(r'([\d.]+)', str(raw).strip())
    if match:
        return float(match.group(1))
    return np.nan

# Build a clean working DataFrame
aerodromes = pd.DataFrame({
    'name':              aerodromes_raw['ICAO_Code'],          # use ICAO as primary label
    'icao':              aerodromes_raw['ICAO_Code'],
    'type':              aerodromes_raw['Type'],
    'traffic':           aerodromes_raw['Traffic_Permitted'],
    'operator':          aerodromes_raw['Admin_Company'],
    'latitude':          aerodromes_raw['latitude'],
    'longitude':         aerodromes_raw['longitude'],
    'elevation_raw':     aerodromes_raw['Elevation_m_ft'],
    'elevation_ft':      aerodromes_raw['Elevation_m_ft'].apply(parse_elevation_ft),
    'remarks':           aerodromes_raw['Remarks_NonSpatial'],
})

print("\nCleaned DataFrame:")
aerodromes.head()
```

### 2.4 UNDERSTANDING OUR DATA


```python
print("Column names:")
print(aerodromes.columns.tolist())

print("\nData types:")
print(aerodromes.dtypes)

print("\nNumerical summary:")
aerodromes[['latitude', 'longitude', 'elevation_ft']].describe()
```

### 4.1 IDENTIFYING ISOLATED AERODROMES


```python
from scipy.spatial import KDTree

coords = aerodromes[['latitude', 'longitude']].dropna().values
icaos  = aerodromes.loc[aerodromes[['latitude', 'longitude']].dropna().index, 'icao'].values

tree = KDTree(coords)
distances, _ = tree.query(coords, k=2)  # k=2: self + nearest neighbor
nearest_distances = distances[:, 1] * 111  # rough degrees-to-km

isolation_analysis = pd.DataFrame({
    'icao': icaos,
    'nearest_distance_km': nearest_distances
}).sort_values('nearest_distance_km', ascending=False)

print("TOP 10 MOST ISOLATED AERODROMES:")
isolation_analysis.head(10)
```

```python
# Visualize isolated aerodromes
plt.figure(figsize=(12, 9))

plt.scatter(aerodromes['longitude'], aerodromes['latitude'],
            alpha=0.3, s=30, c='gray', label='All Aerodromes')

top_isolated = isolation_analysis.head(5)
for _, row in top_isolated.iterrows():
    airport = aerodromes[aerodromes['icao'] == row['icao']].iloc[0]
    plt.scatter(airport['longitude'], airport['latitude'],
                s=200, c='red', edgecolors='black', linewidth=2, zorder=5)
    plt.annotate(f"{airport['icao']}\n({row['nearest_distance_km']:.0f} km to next)",
                 (airport['longitude'], airport['latitude']),
                 xytext=(10, 10), textcoords='offset points', fontsize=8,
                 bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7))

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Most Isolated Aerodromes in Kenya', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### 4.2 ELEVATION ZONES ANALYSIS


```python
def elevation_zone(elevation):
    if pd.isna(elevation):       return 'Unknown'
    elif elevation < 1000:        return 'Lowland (<1000 ft)'
    elif elevation < 3000:        return 'Midland (1000-3000 ft)'
    elif elevation < 5000:        return 'Highland (3000-5000 ft)'
    elif elevation < 7000:        return 'Mountain (5000-7000 ft)'
    else:                         return 'Very High (>7000 ft)'

aerodromes['elevation_zone'] = aerodromes['elevation_ft'].apply(elevation_zone)

zone_counts = aerodromes['elevation_zone'].value_counts()
print("Aerodromes by elevation zone:")
print(zone_counts)

palette = ['#4FC3F7', '#81C784', '#FFD54F', '#FF8A65', '#E57373']
plt.figure(figsize=(9, 7))
plt.pie(zone_counts.values, labels=zone_counts.index, autopct='%1.1f%%',
        startangle=90, colors=palette)
plt.title('Aerodromes by Elevation Zone', fontsize=14, fontweight='bold')
plt.axis('equal')
plt.show()
```

### 5.1 CREATING BUFFER ZONES


```python
# In aviation, buffer zones are used for obstacle limitation surfaces and noise monitoring.

def create_buffer_rectangle(center_lat, center_lon, size_km=10):
    """Return a rectangular bounding box around a centre point."""
    lat_offset = size_km / 111
    lon_offset = size_km / (111 * abs(math.cos(math.radians(center_lat))))
    return {
        'min_lat': center_lat - lat_offset,
        'max_lat': center_lat + lat_offset,
        'min_lon': center_lon - lon_offset,
        'max_lon': center_lon + lon_offset
    }

# 20 km buffer around JKIA
jkia = aerodromes[aerodromes['icao'] == 'HKJK'].iloc[0]
buffer_20km = create_buffer_rectangle(jkia['latitude'], jkia['longitude'], 20)

print(f"20 km buffer around {jkia['icao']}:")
print(f"  Latitude:  {buffer_20km['min_lat']:.4f} → {buffer_20km['max_lat']:.4f}")
print(f"  Longitude: {buffer_20km['min_lon']:.4f} → {buffer_20km['max_lon']:.4f}")

in_buffer = aerodromes[
    (aerodromes['latitude']  >= buffer_20km['min_lat']) &
    (aerodromes['latitude']  <= buffer_20km['max_lat']) &
    (aerodromes['longitude'] >= buffer_20km['min_lon']) &
    (aerodromes['longitude'] <= buffer_20km['max_lon']) &
    (aerodromes['icao'] != 'HKJK')
]

print(f"\nAerodromes within 20 km of JKIA: {len(in_buffer)}")
in_buffer[['icao', 'traffic', 'elevation_ft']]
```

```python
# Visualize the JKIA buffer
from matplotlib.patches import Rectangle

plt.figure(figsize=(10, 9))

plt.scatter(aerodromes['longitude'], aerodromes['latitude'],
            alpha=0.4, s=40, c='gray', label='Other Aerodromes')

plt.scatter(jkia['longitude'], jkia['latitude'],
            s=250, c='red', marker='*', edgecolors='black', linewidth=2,
            label=f"{jkia['icao']} (JKIA)", zorder=5)

if len(in_buffer) > 0:
    plt.scatter(in_buffer['longitude'], in_buffer['latitude'],
                s=120, c='yellow', edgecolors='black', linewidth=1,
                label=f'Within 20 km ({len(in_buffer)})', zorder=4)

rect = Rectangle(
    (buffer_20km['min_lon'], buffer_20km['min_lat']),
    buffer_20km['max_lon'] - buffer_20km['min_lon'],
    buffer_20km['max_lat'] - buffer_20km['min_lat'],
    linewidth=2, edgecolor='red', facecolor='none', linestyle='--',
    label='20 km Buffer Zone'
)
plt.gca().add_patch(rect)

# Label all aerodromes
for _, row in aerodromes.iterrows():
    plt.annotate(row['icao'], (row['longitude'], row['latitude']),
                 xytext=(4, 4), textcoords='offset points', fontsize=7)

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title(f'20 km Buffer Zone Around JKIA', fontsize=14, fontweight='bold')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### 5.2 HAVERSINE DISTANCE — Nearest Other Aerodrome to JKIA


```python
def haversine_distance(lat1, lon1, lat2, lon2):
    """Great-circle distance in km between two lat/lon points."""
    R = 6371
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    dlat, dlon = lat2 - lat1, lon2 - lon1
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    return R * 2 * math.asin(math.sqrt(a))

jkia_lat, jkia_lon = jkia['latitude'], jkia['longitude']

distances = []
for _, row in aerodromes.iterrows():
    if row['icao'] != 'HKJK' and pd.notna(row['latitude']):
        dist = haversine_distance(jkia_lat, jkia_lon, row['latitude'], row['longitude'])
        distances.append({'icao': row['icao'], 'distance_km': round(dist, 1), 'traffic': row['traffic']})

distance_df = pd.DataFrame(distances).sort_values('distance_km')
print("Nearest aerodromes to JKIA:")
distance_df
```

### 5.3 CLUSTER ANALYSIS WITH DBSCAN


```python
from sklearn.cluster import DBSCAN

coords_cluster = aerodromes[['latitude', 'longitude']].dropna().values
clustering = DBSCAN(eps=0.5, min_samples=2).fit(coords_cluster)

aerodromes_clustered = aerodromes.loc[aerodromes[['latitude', 'longitude']].dropna().index].copy()
aerodromes_clustered['cluster'] = clustering.labels_

n_clusters = len(set(clustering.labels_)) - (1 if -1 in clustering.labels_ else 0)
n_noise    = list(clustering.labels_).count(-1)

print(f"Found {n_clusters} cluster(s) of aerodromes")
print(f"{n_noise} aerodrome(s) are isolated (not in any cluster)")

print("\nAerodromes by cluster:")
for cluster_id in sorted(set(clustering.labels_)):
    label = 'Isolated' if cluster_id == -1 else f'Cluster {cluster_id}'
    members = aerodromes_clustered[aerodromes_clustered['cluster'] == cluster_id]
    print(f"\n{label} ({len(members)} aerodromes):")
    for _, row in members.iterrows():
        print(f"  - {row['icao']}")
```

```python
# Visualize clusters
plt.figure(figsize=(12, 9))

unique_clusters = sorted(set(clustering.labels_))
colors_map = plt.cm.tab10(np.linspace(0, 1, len(unique_clusters)))

for cluster_id, color in zip(unique_clusters, colors_map):
    subset = aerodromes_clustered[aerodromes_clustered['cluster'] == cluster_id]
    label  = 'Isolated' if cluster_id == -1 else f'Cluster {cluster_id}'
    c_col  = 'gray' if cluster_id == -1 else color
    plt.scatter(subset['longitude'], subset['latitude'],
                c=[c_col], s=80, label=label,
                edgecolors='black', linewidth=0.6, alpha=0.8)

for _, row in aerodromes_clustered.iterrows():
    plt.annotate(row['icao'], (row['longitude'], row['latitude']),
                 xytext=(4, 4), textcoords='offset points', fontsize=7)

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Aerodrome Clusters (DBSCAN, eps=0.5°≈55 km)', fontsize=14, fontweight='bold')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
